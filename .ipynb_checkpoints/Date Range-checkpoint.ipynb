###                                      Date_Range Function: If Dates are missing

```python
import pandas as pd
```

```python
data=pd.read_csv("AppleHistData.csv")
```

```python
data.shape
```

```python
data.head()
```

```python
rang=pd.date_range(start="25-10-2016",end="19-08-2021",freq="B")  # freq is used to for business days i.e exclude holidays
```

```python
data['Date']=rang
```

```python
rang
```

```python
data.set_index('Date',inplace=True)   # It is used to set the index
```

```python
data
```

```python
%matplotlib inline
```

```python
data['Close'].resample("M").mean().plot()
```
